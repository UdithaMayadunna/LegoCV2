## Get URL

```python
import requests
from bs4 import BeautifulSoup

# URL
url = 'https://www.ebay.com/sch/i.html?_dkr=1&iconV2Request=true&_blrs=recall_filtering&_ssn=weyo-1137&_oac=1&_nkw=lego+minifigure&_sop=15&_ipg=240'

# HTTP 
response = requests.get(url)

if response.status_code == 200:
    
    soup = BeautifulSoup(response.content, 'html.parser')

    # find URL
    product_urls = [item.get('href') for item in soup.find_all('a', class_='s-item__link')]

    # print URL
    for url in product_urls:
        print(f"'{url}',")
else:
    print("can not find")
```
